# Exercise 2: Vegetation Index Analysis
In this exercise, you will analyze vegetation changes within one growing season in a selected agricultural region (Kujawy, Poland).

## Task 1: Define AOI
Create an Area of Interest (AOI) using the coordinates provided below.

```python
import ee
import geemap
#ee.Authenticate()
ee.Initialize(project='ee-dabrowaremote')
```

```python
# TODO: Define AOI
north = 50.4000
south = 50.3000
west = 18.8200
east = 18.9600
aoi = ee.Geometry.Rectangle([west, south, east, north])
```

## Task 2: Retrieve Sentinel-2 Imagery
Retrieve two Sentinel-2 satellite images from the following dates: 2023-04-01 and 2023-07-01.

```python
# TODO: Retrieve Sentinel-2 images for the selected dates.
sentinel2 = ee.ImageCollection('COPERNICUS/S2_HARMONIZED').filterBounds(aoi).filterDate('2023-04-01', '2023-07-01').sort('CLOUDY_PIXEL_PERCENTAGE')
sentinel2


```

```python
image = sentinel2.first()
true_color_params = {'bands': ['B4', 'B3', 'B2'], 'min': 0, 'max': 3000}
sentinel2_june = sentinel2.filter(ee.Filter.calendarRange(6, 6, 'month'))
image2 = sentinel2_june.first()
m = geemap.Map(center=[50.35, 18.89], zoom=10)
m.addLayer(image, true_color_params, "April")
m.addLayer(image2, true_color_params, "June")
m
```

## Task 3: Calculate NDVI
Calculate the Normalized Difference Vegetation Index (NDVI) for both images.

```python
# TODO: Calculate NDVI for both images (Choose the appropriate S-2 bands).
ndvi_april = image.normalizedDifference(['B8', 'B4']).rename('NDVI_April')
ndvi_june = image2.normalizedDifference(['B8', 'B4']).rename('NDVI_June')
```

## Task 4: Compare NDVI
Display both NDVI indices on the map and calculate and visualize their difference.

```python
# TODO: Compare NDVI, visualize the differences, and analyze vegetation changes.

m = geemap.Map(center=[50.35, 18.89], zoom=10)
ndvi_params = {'min': -1, 'max': 1, 'palette': ['blue', 'white', 'green']}
m.addLayer(ndvi_april, ndvi_params, "April")
m.addLayer(ndvi_june, ndvi_params, "June")
m
```
